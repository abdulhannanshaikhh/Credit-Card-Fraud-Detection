```python
#Importing Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading the credit card data
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
#Printing first five rows of the dataset
credit_card_data.head()
```

```python
#Printing the last five rows of the dataset
credit_card_data.tail()
```

```python
#getting dataset information
credit_card_data.info()
```

```python
#checking the number of missing values
credit_card_data.isnull().sum()
```

```python
#Distribution of legit transactions and fraudulent transaction where 0->legit transaction and 1->fraudulent transaction
credit_card_data['Class'].value_counts()
```

```python
#seperating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
#Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare tlegit.Amount.describe()he values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
#Technique=Under-Sampling ->Build a sample dataset containing similar distribution of normal and fraud transaction
#Number of fraud transaction=492
```

```python
legit_sample=legit.sample(n=492)
```

```python
#Creating new dataset by Concatenating two data frames
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
#Printing first five rows of the new dataset
new_dataset.head()
```

```python
#Printing last five rows of the new dataset
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Split data into featues and targets
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Spliting the data into training data and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Training the model
model=LogisticRegression()
```

```python
#training the logistic Regression model with the training data
model.fit(X_train, Y_train)
```

```python
#Finding the performance / Evaluation of the model
#Accuracy Score of training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy of the training data: ',training_data_accuracy)
```

```python
#Accuracy Score of testing data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy of the testing data: ',testing_data_accuracy)
```
